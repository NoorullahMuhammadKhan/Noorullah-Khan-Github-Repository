The Yelp dataset is a collection of user-generated reviews and associated data for businesses in various cities. This data can be used to analyze patterns and trends in consumer behavior, business performance, and geographic locations. * __business_id__ - A unique identifier for each business in the dataset

 # Q1. Check and remove missing data
 
### Q1.1 Write a Python code snippet that checks for missing values in each column of the dataset. If any, only display the names of the columns with missing values and their corresponding number of missing values. Print the length of the data before removing the missing data.

Note: Only output the number of missing values for the columns having at least one missing value!

```python
import pandas as pd
import random
import numpy as np
df = pd.read_csv('Yelp_Portfolio1_Input.csv') #imports the data into the Project
df.head() # Shows first 5 rows of the table
```

```python
# Length of data before removing missing data
len(df)
```

```python

# Check for missing values in each column
missing_values = df.isnull().sum()

# Display the names of the columns with missing values and their corresponding number of missing values
missing_columns = missing_values[missing_values > 0]
if not missing_columns.empty:
    print("Columns with missing values:")
    for column_name, num_missing in missing_columns.iteritems():
        print(f"{column_name}: {num_missing}")

# Print the length of the data before removing the missing data
print(f"Length of data before removing missing values: {len(df)}")
```

### Q1.2 Remove any row that contains at least one missing value, and output the length of the resulting cleaned dataset; After that, remove any row that contain invalid value with either "#NAME?" or "#VALUE!" in the `business_id` and `user_id` columns, and output the length of the resulting cleaned dataset.

```python
# Removing missing values from dataset of missing values
df_cleaned = df.dropna()

# Print length of cleaned dataset
print("Cleaned Dataset:", len(df_cleaned))
```

```python
# Remove rows with invalid values
df_valcleaned = df_cleaned[~df_cleaned['business_id'].isin(['#NAME?', '#VALUE!']) & ~df_cleaned['user_id'].isin(['#NAME?', '#VALUE!'])]

# Length of the resulting value cleaned dataset
print("Values Cleaned Dataset:")
print(len(df_valcleaned))
```

# Q2. Random Subset Selection of Yelp Businesses by City

Selecting a random subset of cities from the Yelp business dataset and extracting all the rows corresponding to businesses located in those cities can be useful for various purposes. For example, it can be used to 
perform exploratory data analysis on a smaller subset of the dataset, which can be more manageable and faster to process than the entire dataset. Suppose you want to select a random subset of 10 cities from the dataset and extract all the rows that correspond to businesses located in those cities. At last, print the length of the resulting sample data. Write Python code that accomplishes this task.

Note: Use the $random.sample()$ function to select 10 random cities from the list of unique cities. Set the random seed to 42 before selecting the cities!!!

```python
# Set the random seed
random.seed(42)

# Get the unique cities and select random 10 cities
unique_cities = df['business_city'].unique()
selected_cities = random.sample(list(unique_cities), 10)

# Filter the dataframe to include only rows with businesses located in the selected cities
sample_data = df[df['business_city'].isin(selected_cities)]

# Print the length of the resulting sample data
print("Length of Resulting Sample Data:")
print(len(sample_data))

# Print the Dataframe
print(sample_data)
```

# Q3. Descriptive statistics on the data output from Q2
### Q3.1 Compute summary statistics for the `stars` column of the sample data
Note: the resulting output includes the count, mean, standard deviation, minimum, maximum values of the column.

```python
# Compute summary statistics for the "stars" column
stars_summary = df['stars'].describe()

# Print summary statistics
print("Summary statistics for the 'stars' column:")
print(stars_summary)
```

### Q3.2 For each city in the dataframe, how many unique businesses are there?

Note: the resulting dataframe has two columns: `business_city` and `count`. Compute summary statistics (similar as Q 3.1) for the `count` column.

```python
# Counting unique businesses by city by grouping 'business_city', 'count', and 'business_id'
city_business_count = df.groupby('business_city')['business_id'].nunique().reset_index()
# (Re)name resulting dataframe
city_business_count.columns = ['business_city', 'count']
# Generating summary statistics
count_summary = city_business_count['count'].describe()
# Print the summary statistics
print(count_summary)
```

### Q3.3 For each business category and business ID combination in the dataframe, how many unique users have rated the business?

Note: the resulting dataframe has three columns: `business_categories`, `business_id`, and `count`. Compute summary statistics (similar as Q 3.1) for the `count` column.

```python
# Group by business_categories' and 'business_id' columns, count unique 'user_id' for each group
business_user_count = df.groupby(['business_categories', 'business_id'])['user_id'].nunique().reset_index()

# (Re)name resulting dataframe
business_user_count.columns = ['business_categories', 'business_id', 'count']

# Compute summary statistics
summary_stats = business_user_count['count'].describe()

# Print the summary statistics
print(summary_stats)
```

## Analysis of Yelp Data
### Distribution of Data

It can be seen that the sampled data provides an accurate representation of the true trends shown in the total (cleaned data). This is shown in all the graphs, and also in the key statistics. However, the sampled data does simplify the total data. This can best be seen in the box plot. There are many small ups and downs shown in the box plot of the cleaned data. However, these are almost all gone when looking at the sampled data, where it seems to show it all almost uniform with the exception of the Saguaro Lake city which is shown a lower. It can be seen in the key statistics that the mean for stars in the sample data is 3.711 (2dp). This is less than 1/10th off from the mean in the cleaned data which is 3.769 (2dp). This is a miniscule difference. The min, max, Q1, Q2 and Q3 were both the same. Therefore, it can be accepted that the sample data is an accurate and unbiased representation of the overall data.

A left-skewed distribution is observed for the stars rating, which indicates that most businesses tend to have higher ratings. This could be because customers are generally satisfied with the services provided, or they're more likely to leave a review when they have a positive experience.

In the scatterplot comparing the votes to the number of stars, it can be seen that more useful votes are 5 star, than oter stars. In the sampled data it can also be seen that there are a slightly more 1 star votes than 2-3 star votes. This could be due to polarisation, and people are more pushed to mark a review if its on either extreme end of the spectrum, such as 1 and 5. This means if something is extraordinarily good, such as 5 star, or if something is rather awful, deserving a 1 star it will be found helpful by more people.

It can be seen in the histogram looking at distribution of stars, that there are substantially more 5 star reviews than other reviews. This is also supported in both Box Plots.

This leads to an inflation of 5 stars, and the star system becoming slighly less reliable and valid, because every user does not leave a valid review, and it's more likely for a user to leave a 5 star review.

The box plots also show that the majority of stars, Q1 - Q3 are in the 3 - 5 range.
Therefore, it can be concluded that users are more likely to leave a positive review.
